```python
def plot_outliers(dataframe):
    # Flatten the DataFrame for box plot
    df_long = pd.melt(dataframe)

    # Create a box plot
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='variable', y='value', data=df_long)

    # Show the plot
    plt.title('Box Plot with Outliers for All Columns')
    plt.xticks(rotation=45)
    plt.show()

    # Adjust layout
    plt.tight_layout()
    plt.show()
```

```python
def plot_individual_boxplots(dataframe):
    # Set up subplots
    fig, axes = plt.subplots(nrows=len(dataframe.columns), figsize=(8, 4 * len(dataframe.columns)))

    # Iterate through columns
    for i, column_name in enumerate(dataframe.columns):
        # Create a box plot
        sns.boxplot(x=dataframe[column_name], ax=axes[i])

        # Set plot title
        axes[i].set_title(f'Box plot - {column_name}')

    # Adjust layout
    plt.tight_layout()
    plt.show()
```

```python
plot_outliers(df)
plot_individual_boxplots(df)
```

```python
def plot_correlation_matrix(dataframe):
    # Calculate the correlation matrix
    corr_matrix = dataframe.corr()

    # Create a heatmap
    plt.figure(figsize=(25, 20))
    sns.heatmap(corr_matrix, annot=True,  fmt=".2f", linewidths=.5)

    # Show the plot
    plt.title('Correlation Matrix')
    plt.show()
```

```python
plot_correlation_matrix(df)
```

```python
def plot_histograms(dataframe):
    # Set up subplots
    fig, axes = plt.subplots(nrows=len(dataframe.columns), ncols=1, figsize=(8, 4 * len(dataframe.columns)))

    # Iterate through columns
    for i, column_name in enumerate(dataframe.columns):
        # Create a histogram
        dataframe[column_name].hist(ax=axes[i], bins=20, edgecolor='black', color='#307D7E')

        # Set plot title and labels
        axes[i].set_title(f'Histogram - {column_name}')
        axes[i].set_xlabel(column_name)
        axes[i].set_ylabel('Frequency')

    # Adjust layout
    plt.tight_layout()
    plt.show()
```

```python
plot_histograms(df)
```
